```python
# import relevant libraries
import pandas as pd
import numpy as np
```

```python
# read dataset into dataFrame, overview of dataset
strokeData = pd.read_csv('cleanedDataset.csv')
strokeData.info()
```

```python
strokeData.head()
```

```python
# defining features and target
X = strokeData.drop(columns='stroke', axis=1)
y = strokeData['stroke']
```

```python
y
```

```python
# check imblearn version
import imblearn
print(imblearn.__version__)
```

```python
# import sklearn libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, recall_score
```

```python
# split LR model: 70-30 (imbalanced)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
# import SMOTE
from imblearn.over_sampling import RandomOverSampler
```

```python
# oversample target variable
oversample = RandomOverSampler(sampling_strategy='minority')
X_oversample, y_oversample = oversample.fit_resample(X_train, y_train)
print(y_oversample.value_counts())
```

```python
y_test.value_counts()
```

```python
y_train.value_counts()
```

```python
# import sklearn library - K-Nearest Neighbours
from sklearn.neighbors import KNeighborsClassifier
```

```python
# create Classifier for Logistic Regression - logreg
oversample_knn = KNeighborsClassifier(n_neighbors=60)
```

```python
# fit classifier into oversampled training data
oversample_knn.fit(X_oversample, y_oversample)
```

```python
# predict labels of oversampled test dataset
y_pred = oversample_knn.predict(X_test)
```

```python
# compute and print oversampled confusion matrix
print(confusion_matrix(y_test, y_pred))
```

```python
# compute and print oversampled classification report
print(classification_report(y_test, y_pred, target_names=['noStroke', 'stroke']))
```

```python
# recall rate
recall = recall_score(y_test, y_pred)
recall
```

```python
print(f'K-Nearest Neighbours Oversampling Recall Rate: {recall*100:.2f}%')
```

Recursive Feature Elimination - find features that correlate with stroke

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
```

```python
model = LinearRegression()

#Initializing RFE model
rfe = RFE(model, 10)

#Transforming data using RFE
X_rfe = rfe.fit_transform(X,y)

#Fitting the data to model
model.fit(X_rfe,y)
print(rfe.support_)
print(rfe.ranking_)
```

```python
#no of features
nof_list=np.arange(1,13)            
high_score=0

#Variable to store the optimum features
nof=0           
score_list =[]
for n in range(len(nof_list)):
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 42)
    model = LinearRegression()
    rfe = RFE(model,nof_list[n])
    X_train_rfe = rfe.fit_transform(X_train,y_train)
    X_test_rfe = rfe.transform(X_test)
    model.fit(X_train_rfe,y_train)
    score = model.score(X_test_rfe,y_test)
    score_list.append(score)
    if(score>high_score):
        high_score = score
        nof = nof_list[n]

print("Optimum number of features: %d" %nof)
print("Score with %d features: %f" % (nof, high_score))
```

```python
cols = list(X.columns)
model = LinearRegression()

#Initializing RFE model
rfe = RFE(model, 10)
#Transforming data using RFE
X_rfe = rfe.fit_transform(X,y)

#Fitting the data to model
model.fit(X_rfe,y)              
temp = pd.Series(rfe.support_,index = cols)
selected_features_rfe = temp[temp==True].index
print(selected_features_rfe)
```
