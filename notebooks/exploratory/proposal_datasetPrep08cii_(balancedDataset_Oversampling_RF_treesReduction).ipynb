```python
# import relevant libraries
import pandas as pd
import numpy as np
```

```python
# read dataset into dataFrame, overview of dataset
strokeData = pd.read_csv('cleanedDataset.csv')
strokeData.info()
```

```python
strokeData.head()
```

```python
# defining features and target
X = strokeData.drop(columns='stroke', axis=1)
y = strokeData['stroke']
```

```python
X
```

```python
y
```

```python
# import sklearn libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, recall_score
```

```python
# import SMOTE
from imblearn.over_sampling import RandomOverSampler
```

```python
# split data 70-30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
# oversample target variable
oversample = RandomOverSampler(sampling_strategy='minority')
X_oversample, y_oversample = oversample.fit_resample(X_train, y_train)
print(y_oversample.value_counts())
```

```python
from math import sqrt
sqrt(3403)
```

```python
# import sklearn library - RF
from sklearn.ensemble import RandomForestClassifier
```

```python
# create Classifier for Random Forest - rf
oversample_rf = RandomForestClassifier(n_estimators=10, min_samples_leaf=34)
```

```python
# fit classifier into oversampled training data
oversample_rf.fit(X_oversample, y_oversample)
```

```python
# predict labels of oversampled test dataset
y_pred = oversample_rf.predict(X_test)
```

```python
# compute and print oversampled confusion matrix
print(confusion_matrix(y_test, y_pred))
```

```python
# compute and print oversampled classification report
print(classification_report(y_test, y_pred, target_names=['noStroke', 'stroke']))
```

```python
# recall score
recall = recall_score(y_test, y_pred)
print(recall)
```

```python
print(f"Random Forest Oversampling Recall Rate: {recall*100:.2f}%")
```
