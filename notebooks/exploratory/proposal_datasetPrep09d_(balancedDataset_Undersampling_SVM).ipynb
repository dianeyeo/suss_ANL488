```python
# import relevant libraries
import pandas as pd
import numpy as np
```

```python
# read dataset into dataFrame, overview of dataset
strokeData = pd.read_csv('cleanedDataset.csv')
strokeData.info()
```

```python
strokeData.head()
```

```python
# defining features and target
X = strokeData.drop(columns='stroke', axis=1)
y = strokeData['stroke']
```

```python
X
```

```python
y
```

```python
# import sklearn libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, recall_score
```

```python
# import SMOTE
from imblearn.under_sampling import RandomUnderSampler
```

```python
# split data 70-30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# oversample target variable
undersample = RandomUnderSampler()
X_undersample, y_undersample = undersample.fit_resample(X_train, y_train)
print(y_undersample.value_counts())
```

```python
# import sklearn library - SVM
from sklearn.svm import SVC
```

```python
# create Classifier for Support Vector Machine - svm
undersample_svm = SVC(kernel='linear')
```

```python
# fit classifier into oversampled training data
undersample_svm.fit(X_undersample, y_undersample)
```

```python
# predict labels of oversampled test dataset
y_pred = undersample_svm.predict(X_test)
```

```python
# compute and print oversampled confusion matrix
print(confusion_matrix(y_test, y_pred))
```

```python
# compute and print oversampled classification report
print(classification_report(y_test, y_pred, target_names=['noStroke', 'stroke']))
```

```python
# recall score
recall = recall_score(y_test, y_pred)
print(recall)
```

```python
print(f"Support Vector Undersampling recall rate: {recall*100:.2f}%")
```
